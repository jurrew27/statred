```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.mlab as ml

class NBClassifier:
    def __init__(self):
        self.pdfsMale = [] 
        self.pdfsFemale = []
        self.meansMale = [] 
        self.meansFemale = []
        self.stdsMale = [] 
        self.stdsFemale = []
    
    def create_normal_pdf(self, mu, sigma):
        def _normal_func(x):
            return 1 / (np.sqrt(2 * np.pi * sigma**2)) * np.exp(-(x - mu)**2 / (2 * sigma**2))
        return _normal_func
    
    def train(self, data, target):
        # Male is X=0, Female is X=1
        indicesMale = np.where(target == 0)[0]
        indicesFemale = np.where(target == 1)[0]
        
        for i in range(len(data[0])):
            mean = np.mean(data[indicesMale, i])
            std = np.std(data[indicesMale, i])
            self.meansMale.append(mean)
            self.stdsMale.append(std)
            self.pdfsMale.append(self.create_normal_pdf(mean, std))
            
            mean = np.mean(data[indicesFemale, i])
            std = np.std(data[indicesFemale, i])
            self.meansFemale.append(mean)
            self.stdsFemale.append(std)
            self.pdfsFemale.append(self.create_normal_pdf(mean, std))
            
    def predict(self, data):
        maleProb = femaleProb = 0
        for i in range(len(data)):
            maleProb += self.pdfsMale[i](data[i])
            femaleProb += self.pdfsFemale[i](data[i])
        maleProb /= len(data)
        femaleProb /= len(data)
        
        return maleProb < femaleProb
```

```python
data = ml.csv2rec('biometrie2014.csv', delimiter=',', skiprows=4)
data = np.array(list(map(list, data)))
# Male is X=0, Female is X=1
target = np.array(list(map(lambda x: 0 if x == 'M ' else 1, data[:, 0])))
data = np.delete(data, 0, axis=1).astype(np.float)

bayes = NBClassifier()
bayes.train(data, target)

rvs = ['Weight', 'Length', 'Shoesize']

print('RV || mean male | std male || mean female | std female')
for (i, val) in enumerate(rvs):
    print(val + ' || ' + str(bayes.meansMale[i]) + ' | ' + str(bayes.stdsMale[i]) + \
          ' || ' + str(bayes.meansFemale[i]) + ' | ' + str(bayes.stdsFemale[i]))

xs = [np.linspace(0, 150, 100), np.linspace(125, 225, 100), np.linspace(30, 50, 100)]
for (i, val) in enumerate(rvs):
    plt.figure()
    plt.title(val)
    plt.plot(xs[i], bayes.pdfsMale[i](xs[i]), 'b-')
    plt.plot(xs[i], bayes.pdfsFemale[i](xs[i]), 'r-')
    plt.show()
```

```python
predicted = [bayes.predict(dataPoint) for dataPoint in data]
predicted = np.array(predicted).astype(int)

cm = np.zeros((2, 2))
# Zeros to ones and ones to zeros or the matrix would be upside down and FLIPped
for a, p in zip(1 - target, 1 - predicted):
    cm[a][p] += 1
        
# [[True positives, False negatives]
# [False positives, True negatives]]
print(cm)
```
