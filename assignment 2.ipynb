```python
#3.3.2
import numpy as np
import matplotlib.pyplot as plt

x1 = np.random.uniform(low=-1,high=1,size=10000)
x2 = np.random.uniform(low=-1,high=1,size=10000)

plt.figure()
plt.plot(x1, x2, 'r.')
plt.xlabel('X1')
plt.ylabel('X2')
plt.show()
```

```python
#3.4.1
def standard_normal_def(x):
    return 1 / (np.sqrt(2 * np.pi)) * np.exp(-(x * x) / 2.0)
```

```python
#3.4.2
def normal_pdf(x, mu, sigma):
    return standard_normal_def((x - mu) / np.sqrt(sigma)) / np.sqrt(sigma)
```

```python
#3.4.3
import scipy.special

def standard_normal_cdf(x):
    return 0.5 * (1 + scipy.special.erf(x / np.sqrt(2)))

def normal_cdf(x, mu, sigma):
    return standard_normal_cdf((x - mu) / np.sqrt(sigma))
                                                                   
```

```python
#3.4.4
x = np.linspace(-2, 8, 100)
plt.figure()
plt.plot(x, normal_pdf(x, 3, 4), 'b-')
plt.plot(x, normal_cdf(x, 3, 4), 'r-')
plt.show()
```

```python
#3.4.5
values = np.random.normal(loc=3, scale=2, size=10000)
bins = np.linspace(values.min(), values.max(), 50)
h, _ = np.histogram(values, bins=bins, density=True)
x = np.linspace(-5, 10, 100)

plt.figure()
plt.bar(bins[:-1], h, bins[1] - bins[0], color='g')
plt.plot(x, normal_pdf(x, 3, 4), 'r-')
plt.show()
```

```python
#3.4.6
values = np.random.uniform(low=-1,high=1,size=10000)
sigma = values.std()
mu = values.mean()
x = np.linspace(-5, 5, 100)
y = normal_pdf(x, mu, sigma)

plt.figure()
plt.plot(x, y, 'r-')
plt.show()
```
